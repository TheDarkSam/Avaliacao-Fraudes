```python
#removendo numeros e caracteres nao latinos do dataframe
df['nome_do_funcionario'] = df['nome_do_funcionario'].str.replace(r'[^a-zA-Z\s]', '', regex=True)
```

```python
#detecta as compras duplicadas
def detectar_compras_duplicadas(compras):
    compras_duplicadas = compras[compras.duplicated(subset=['nome_do_funcionario', 'item_comprado'], keep=False)]
    return compras_duplicadas
```

```python
#faz a verificação de suspeitas, adicionando algumas colunas para ajudar na verificacao
def verificar_valores_suspeitos(compras):
    std = compras['valor_em_real'].std()
    compras['horario_suspeito'] = compras['data_da_compra'].apply(compras_fora_de_horario)
    compras['valor_suspeito'] = compras['valor_em_real'] > 3 * std
    compras['infracao'] = compras.apply(classificar_infracao, axis=1)
    return compras
```

```python
#classifica a infracao como suspeita ou severa
def classificar_infracao(compra):
  if compra['horario_suspeito'] and compra['valor_suspeito']:
    return 'Infração Severa'
  elif compra['horario_suspeito'] or compra['valor_suspeito']:
    return 'Infração Moderada'
  else:
    return 'Nenhuma Infração'
```

```python
#retorna se a compra foi fora do horario comercial
def compras_fora_de_horario(data_da_compra):
    data_da_compra = pd.to_datetime(data_da_compra)
    return (data_da_compra.hour < 8) or (data_da_compra.hour > 18)
```

```python
# cria um dicionario com o nome do comprador, e uma lista com todas as compras dele
def organizar_por_servidor(compras):
    compras = compras.groupby('nome_do_funcionario')['item_comprado'].apply(list).to_dict()
    #df.groupby("FaixaIdade")["Churn"].value_counts(normalize=True)
    return compras
```

```python
def gerar_relatorio(compras):
    #listando compras duplicadas
    compras_duplicadas = detectar_compras_duplicadas(compras)
    print("Compras duplicadas:")
    print(compras_duplicadas)

    #removendo compras duplicadas
    compras = compras.drop_duplicates()

    #listando valores acima de 1000
    print("Valores acima de 1000:")
    print(compras[compras['valor_em_real'] > 1000])

    #verificando compras fora do horário
    compras = verificar_valores_suspeitos(compras)
    print("Compras fora do horário:")
    print(compras[compras['horario_suspeito']])

    #ordedando compras por servidor
    compras_por_servidor = organizar_por_servidor(compras)
    print("Compras organizadas por servidor:")
    print(compras_por_servidor)

    #Exibindo severidade da infração
    compras = verificar_valores_suspeitos(compras)
    print("Infrações encontradas:")
    print(compras['infracao'])
    infracoes_detalhadas = compras.groupby(['infracao', 'nome_do_funcionario']).size().unstack(fill_value=0)

    #exibem a pessoa e quais tipos de infracoes ou suspeita elas tiveram
    print("Infrações detalhadas:")
    print(infracoes_detalhadas)
```
